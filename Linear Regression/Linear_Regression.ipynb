### Learning the model

htwt.txt is data collected from a lot of people at NITK sometime back.
Our task is to learn a linear regression model to predict a person's weight from the height.

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Load the training data: 1st col- ht; 2nd col - weight
x = np.loadtxt('htwt.txt')
plt.plot(x[:,1],x[:,0],'bo')
plt.xlabel('Height -->')
plt.ylabel('Weight -->')



# Plot the height vs weight for all individuals

# Form the matrices H and Y by copying parts of x
# y contains the target values (weight)
# H is the matrix that contains the input (height) and const 1
# H = np.column_stack((x[:, 0], np.ones_like(x[:, 0])))  # [height, 1] matrix

# Use the closed form formula to find the vector a = [m c]^T

# Find the predicted values using the equation y_h = H a


# Plot the predicted values [will be a line]

# Find the Mean Square Error in Train Data


y = x [:, 1]  # Target values
h = np.column_stack((x[:,0],np.ones_like(x[:,0])))

ht = np.transpose(h)
temp = ht @ h
temp2 = np.linalg.inv(temp)
a = (temp2 @ ht) @ y

ycap = h @ a
plt.plot(ycap,x[:,0] , 'r-')

plt.show()


temp3=y-ycap
ME = (temp3.T @ temp3)/len(temp3)
print("Mean Square Error in Train Data: ", ME)
```
